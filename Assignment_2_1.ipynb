2.2 Check for duplicate values and remove them.

```python
import pandas as pd

df = pd.read_csv("Assignment 2 dataset.csv")

print("Number of duplicate rows before removal:", df.duplicated().sum())
df = df.drop_duplicates(inplace=False)
print("Number of duplicate rows after removal:", df.duplicated().sum())
```

2.3 Check for missing values. If a data record (row) only has a few missing values, replace the missing values with the median of the column feature in that specific “Region” variable. If most column values in a data record are missing, remove the data record.

```python
import pandas as pd

df = pd.read_csv("Assignment 2 dataset.csv")
find_missing_values = df.isnull().sum() #check for missing values
print("Number of missing values per column:\n", find_missing_values)


limitPer = len(df.columns) * .51 #anything over 51% is considered "most".
df = df.dropna(thresh=limitPer, axis=1)

df['AveragePrice'] = df['AveragePrice'].fillna(df['AveragePrice'].mean())
df['Total Volume'] = df['Total Volume'].fillna(df['Total Volume'].mean())
df['4046'] = df['4046'].fillna(df['4046'].mean())
df['4225'] = df['4225'].fillna(df['4225'].mean())
df['4770'] = df['4770'].fillna(df['4770'].mean())
df['Total Bags'] = df['Total Bags'].fillna(df['Total Bags'].mean())
df['Small Bags'] = df['Small Bags'].fillna(df['Small Bags'].mean())
df['Large Bags'] = df['Large Bags'].fillna(df['Large Bags'].mean())

print(df)
```

#3. Exploratory Data Analysis (EDA)


3.1 Describe the variables
- Describe all variables in the dataset.
- For continuous variables: report **range (min, max), mean, median, and distribution**.
- For categorical variables: list unique values.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("Assignment 2 dataset.csv")

print("Numerical Variables Summary:")
print(df.describe(include=np.number))

print("\nCategorical Variables Summary:")
print(df.describe(include='object'))

print("\nDetailed statistics for continuous variables")

for col in df.select_dtypes(include=np.number).columns:
    print(f"\nVariable: {col}")
    if not df[col].isnull().all():
        print(f"  Range (min, max): ({df[col].min():.2f}, {df[col].max():.2f})")
        print(f"  Mean: {df[col].mean():.2f}")
        print(f"  Median: {df[col].median():.2f}")

        if df[col].nunique() < 50:
            print(f"  Distribution (Value Counts):\n{df[col].value_counts().sort_index().to_string()}")
        else:
            print(f"  Distribution (first 5 unique values and counts):\n{df[col].value_counts().head().to_string()}")
    else:
        print(f"  Variable {col} has all missing values.")

print("\nDetailed unique values for categorical variables")

for col in df.select_dtypes(include='object').columns:
    print(f"\nVariable: {col}")
    print(f"  Unique values: {df[col].unique()}")
    print(f"  Value counts:\n{df[col].value_counts().to_string()}")
```

3.3 Highest average price
- Find the highest value in "AveragePrice".
- Report which region it belongs to.

```python
import pandas as pd

df = pd.read_csv("Assignment 2 dataset.csv")

highest_value = df['AveragePrice'].max()
print("Highest value in 'AveragePrice':", highest_value)

region_highest_value = df.loc[df['AveragePrice'] == highest_value, 'region'].values[0]
print("Region with Highest Value:")
print(region_highest_value)

```
